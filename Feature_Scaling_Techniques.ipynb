### FEATURE SCALING TECHNIQUES



## Normalization - Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Sample data
df = pd.DataFrame({'Salary': [30000, 50000, 70000, 90000]})

# Apply Min-Max Scaling
scaler = MinMaxScaler()
df['Salary_scaled'] = scaler.fit_transform(df[['Salary']])

print(df)
```

## Standardization (Z-score Normalization)

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Sample data
df = pd.DataFrame({'Score': [55, 85, 70, 90, 60]})

# Apply Z-score normalization
scaler = StandardScaler()
df['Score_standardized'] = scaler.fit_transform(df[['Score']])

print(df)
```

## Robust Scaling

```python
from sklearn.preprocessing import RobustScaler
import pandas as pd

# Sample data with outliers
df = pd.DataFrame({'Income': [25000, 30000, 28000, 500000, 27000]})

# Apply Robust Scaling
scaler = RobustScaler()
df['Income_scaled'] = scaler.fit_transform(df[['Income']])

print(df)
```

```python
import numpy as np
from sklearn.preprocessing import RobustScaler

df['Income_log'] = np.log1p(df['Income'])

scaler = RobustScaler()
df['Income_log_scaled'] = scaler.fit_transform(df[['Income_log']])

print(df)
```

## MaxAbs Scaling

```python
from sklearn.preprocessing import MaxAbsScaler
import pandas as pd

# Sample data
df = pd.DataFrame({'Feature': [-100, -50, 0, 25, 75]})

# Apply MaxAbs Scaling
scaler = MaxAbsScaler()
df['Feature_scaled'] = scaler.fit_transform(df[['Feature']])

print(df)
```
