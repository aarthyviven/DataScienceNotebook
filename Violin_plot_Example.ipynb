### Basic Violin Plot using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(42)
group_a = np.random.normal(50, 10, 100)
group_b = np.random.normal(60, 8, 100)
group_c = np.random.normal(55, 12, 100)

data = [group_a, group_b, group_c]

# Create violin plot
plt.figure()
plt.violinplot(data, showmeans=True)
plt.xticks([1, 2, 3], ['Group A', 'Group B', 'Group C'])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Simple Violin Plot Example')

plt.show()
```

### Basic Violin Plot Using Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
df = pd.DataFrame({
    "Group": ["Group A"] * 100 + ["Group B"] * 100 + ["Group C"] * 100,
    "Value": np.concatenate([
        np.random.normal(50, 10, 100),
        np.random.normal(60, 8, 100),
        np.random.normal(55, 12, 100)
    ])
})

# Violin plot
plt.figure()
sns.violinplot(x="Group", y="Value", data=df)
plt.title("Simple Violin Plot using Seaborn")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

### Specifying Color Palette

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
df = pd.DataFrame({
    "Group": ["Group A"] * 100 + ["Group B"] * 100 + ["Group C"] * 100,
    "Value": np.concatenate([
        np.random.normal(50, 10, 100),
        np.random.normal(60, 8, 100),
        np.random.normal(55, 12, 100)
    ])
})

# Violin plot with different colors
plt.figure()
sns.violinplot(
    x="Group",
    y="Value",
    data=df,
    palette=["skyblue", "lightgreen", "salmon"]
)

plt.title("Violin Plot with Different Colors for Each Group")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

## From Seaborn v0.13+, palette should be used only with hue

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
df = pd.DataFrame({
    "Group": ["Group A"] * 100 + ["Group B"] * 100 + ["Group C"] * 100,
    "Value": np.concatenate([
        np.random.normal(50, 10, 100),
        np.random.normal(60, 8, 100),
        np.random.normal(55, 12, 100)
    ])
})

# Violin plot with different colors
plt.figure()
sns.violinplot(
    x="Group",
    y="Value",
    hue="Group",        # explicitly assign hue
    data=df,
    palette="Set2",
    legend=False        # avoid duplicate legend
)

plt.title("Violin Plot with Different Colors for Each Group")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```

## Explicit Color Mapping

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
np.random.seed(42)
df = pd.DataFrame({
    "Group": ["Group A"] * 100 + ["Group B"] * 100 + ["Group C"] * 100,
    "Value": np.concatenate([
        np.random.normal(50, 10, 100),
        np.random.normal(60, 8, 100),
        np.random.normal(55, 12, 100)
    ])
})

# Violin plot with different colors
palette = {
    "Group A": "#4C72B0",
    "Group B": "#55A868",
    "Group C": "#C44E52"
}

sns.violinplot(
    x="Group",
    y="Value",
    hue="Group",
    data=df,
    palette=palette,
    legend=False
)

plt.title("Violin Plot with Different Colors for Each Group")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.show()
```
